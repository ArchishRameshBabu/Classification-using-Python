```python
import numpy as np
import pandas as pd
from sklearn import linear_model,svm
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
import matplotlib.pyplot as plt
from sklearn.ensemble import GradientBoostingClassifier  #GBM algorithm
from sklearn.neighbors import KNeighborsClassifier
from sklearn import metrics   #Additional scklearn functions
from sklearn.model_selection import GridSearchCV   #Performing grid search
from sklearn.naive_bayes import GaussianNB
```

```python
import pandas as pd
data=pd.read_table("patients.txt", sep="\t")
data.head()
```

```python
## Converting the categorical colums to factors
```

```python
### Factarizing the categortical columns 
data.columns=['a','b','c','d','e','f','target']
cleanup_nums = {'a':     {"vhigh": 4, "high": 3,"med": 2,"low": 1},
                'b': {"vhigh": 4, "high": 3,"med": 2,"low": 1},
                'c': {"5more": 5},
                'd': {"more": 5},
                'e': {"small": 1,"med": 2,"big": 3},
                'f': {"low": 1,"med": 2,"high": 3},
                'target' : {"unacc":1, "acc":2, "good":3, "vgood":4 } 
               }
data.replace(cleanup_nums, inplace=True)
data = data.convert_objects(convert_numeric=True)
```

```python
cols=data.shape[1]
x = data.iloc[:,:cols-1]
x=pd.get_dummies(x)
```

```python
y=data.iloc[:,-1]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(x, y, test_size = 0.2, random_state = 0)
```

### Logistic Regression

```python
clf = LogisticRegression(fit_intercept=True,C=1)
clf.fit(X_train, Y_train)

LR_Accuracy_Test=clf.score(X_test, Y_test)
LR_Accuracy_train=clf.score(X_train, Y_train)

#Logistic_Prediction_Train=clf.predict_proba(Train_Data.iloc[:,:499])[:,1]
#Logistic_Prediction_Test=clf.predict_proba(Test_Data.iloc[:,:499])[:,1]

print("Accuracy of Train Data : ",LR_Accuracy_train)
print("Accuracy of Test Data : ",LR_Accuracy_Test)
```

```python
RF_Model_Base = RandomForestClassifier()
RF_Model_Base.fit(X_train, Y_train)

RF_predictions_Test = RF_Model_Base.predict(X_test)
RF_predictions_Train = RF_Model_Base.predict(X_train)

Train_Accuracy_RF=RF_Model_Base.score(X_train, Y_train)
Test_Accuracy_RF=RF_Model_Base.score(X_test, Y_test)

print("Accuracy of Train Data : ",Train_Accuracy_RF)
print("Accuracy of Test Data : ",Test_Accuracy_RF)
```

```python
#making the instance
KNN_Base = KNeighborsClassifier()

#Learning
KNN_Base.fit(X_train, Y_train)
KNN_Accuracy_Test=KNN_Base.score(X_test, Y_test)
KNN_Accuracy_train=KNN_Base.score(X_train, Y_train)

print("Accuracy of Train Data : ",KNN_Accuracy_train)
print("Accuracy of Test Data : ",KNN_Accuracy_Test)
```

```python
NB = GaussianNB() ### No parameters to tune, works well for test data 
    
NB.fit(X_train,Y_train)

NB_Accuracy_Test=KNN_Base.score(X_test, Y_test)
NB_Accuracy_train=KNN_Base.score(X_train, Y_train)

print("Accuracy of Train Data : ",NB_Accuracy_train)
print("Accuracy of Test Data : ",NB_Accuracy_Test)
    
```

```python
#making the instance
GB_model_Base = GradientBoostingClassifier()

#Learning
GB_model_Base.fit(X_train, Y_train)
GB_Accuracy_Test=GB_model_Base.score(X_test, Y_test)
GB_Accuracy_train=GB_model_Base.score(X_train, Y_train)

print("Accuracy of Train Data : ",GB_Accuracy_train)
print("Accuracy of Test Data : ",GB_Accuracy_Test)
```

### Grid Search

```python
C_value = [0.0001,0.0002,0.0003,0.0004,0.0005,0.0006,0.0007,0.0008,0.0009,0.001,0.01,0.1,1]

grid={"C":C_value, "penalty":["l2","l1"]}

Logistic=LogisticRegression()
Log_GS=GridSearchCV(Logistic,grid,cv=3)
Log_GS.fit(X_train,Y_train)

print("Best parameters after Grid Search ",Log_GS.best_params_)

LR_Accuracy_Test=Log_GS.score(X_test,Y_test)
LR_Accuracy_train=Log_GS.score(X_train,Y_train)

Logistic_Prediction_Train=Log_GS.predict_proba(X_train)[:,1]
Logistic_Prediction_Test=Log_GS.predict_proba(X_test)[:,1]

print("Accuracy on Train Data for Logistic Regression after Grid Search: ",LR_Accuracy_train)
print("Accuracy on Test Data for Logistic Regression after Grid Search:: ",LR_Accuracy_Test)
```

```python
RF = RandomForestClassifier()

param_grid = {'n_estimators': [20,45],
    'max_features': ['auto', 'sqrt'],
    'criterion' :['gini', 'entropy']}

# Create a based model

# Instantiate the grid search model
RF_Grid = GridSearchCV(estimator = RF, param_grid = param_grid,cv = 5, n_jobs = -1, verbose = 2)
RF_Grid.fit(X_train,Y_train)

print("The best parameters are",RF_Grid.best_params_)

RF_Accuracy_Test=RF_Grid.score(X_test,Y_test)
RF_Accuracy_train=RF_Grid.score(X_train,Y_train)

RF_Prediction_Train=RF_Grid.predict_proba(X_train)[:,1]
RF_Prediction_Test=RF_Grid.predict_proba(X_test)[:,1]

print("Accuracy on Train Data for Random Forest after Grid Search: ",RF_Accuracy_train)

print("Accuracy on Test Data for Random Forest after Grid Search:: ",RF_Accuracy_Test)

print("F1 Score",metrics.f1_score(Y_test, RF_Grid.predict(X_test), average='weighted'))

pd.crosstab(Y_test, RF_Grid.predict(X_test), rownames=['Actual'], colnames=['Predicted'])
```
